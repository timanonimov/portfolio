```python
import pandas as pd
#read data
data = pd.read_csv('data (1).csv')
data.head()
```

```python
#check
data.describe()
```

```python
#drop null
data.dropna()
```

```python
#what in California?
data[data['city'] == 'California']
```

```python
#Taco Bell in California
data[(data['city'] == 'California') & (data['name'] == 'Taco Bell')]
```

```python
result = data[
  # (city == 'New York' OR name == 'Taco Bell')
  # AND
  # (NOT (menus.name IN ['Volcano Taco', 'Fresco Soft Taco']))

  ((data['city'] == 'New York') | (data['name'] == 'Taco Bell'))
  &
  (~(data['menus.name'].isin(['Volcano Taco', 'Fresco Soft Taco'])))
]
result
```

```python
#restoraunt with menus currency
result = data[~data['menus.currency'].isna()]
result
```

```python
#how amount of the restaurants in different cities
data['city'].value_counts()
```

```python
#result = Top 5 cities with Taco Bell restaurants
result = data[data['name'] == 'Taco Bell']['city'].value_counts()[:5]
result
```

```python
#Which restaurants are open in October?
#turn the column into a datetime data type
data['dateAdded'] = pd.to_datetime(data['dateAdded'])
# see where the month is october (number 10)
data[data['dateAdded'].dt.month == 10]
```

```python
#how many restaurants are open each month?
result = data.groupby(data['dateAdded'].dt.month).agg({'id': 'nunique'})
result
```

```python
#where can i eat pizza?
data[data['categories'].str.contains('Pizza')]
```

```python
#records that have more than 21 categories, then they need to be grouped by provinces and calculate the minimum longitude in each. Round longitude values ​​to 3 decimal places.
tmp = data[data['categories'].str.split(',').str.len() > 20].groupby('province')['longitude'].min().round(3)
#and save it :)
tmp.to_csv('res.csv', sep=';')
```
